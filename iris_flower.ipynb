```python
# import necessary libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# set data visualization and random seed

np.random.seed(0)
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (15, 10)
```

```python
# load the iris dataset from the scikit-learn library

from sklearn.datasets import load_iris

iris_data = load_iris() 
```

```python
# create a dataframe with the feature variables

df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
```

```python
# overview of data

print(df.head())
print(df.info())
print(df.describe())
print(df.shape)
```

```python
# create a new column for the species name

df['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)
df.head()
```

```python
# create test and train data

df['is_train'] = np.random.uniform(0, 1, len(df)) <= .75
df.head()
```

```python
# create dataframes with trainig and test rows

train, test = df[df['is_train']==True], df[df['is_train']==False]
```

```python
# show the number of observations for the training and test dataframes

print('Number of observations in the training data: ', len(train))
print('Number of observations in the test data: ', len(test))
```

```python
# create a list of the feature columns names

features = df.columns[:4]
features
```

```python
# convert each species nameinto digits

y = pd.factorize(train['species'])[0]
y
```

```python
# create a random forest classifier

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_jobs=2, random_state=0)
```

```python
# train the classifier

clf.fit(train[features], y)
```

```python
# apply the trained classifier to the test

clf.predict(test[features])
```

```python
# view the predicted probabilities of the first 10 observations

clf.predict_proba(test[features])[0:10]
```

```python
# map names for the plants for each predicted plant class

pred = iris_data.target_names[clf.predict(test[features])] #this code can also be used to test labelled data to check for accuracy
pred[0:5]
```

```python
# view the ACTUAl species for the first five observations

test['species'].head()
```

```python
# create confusion matrix
pd.crosstab(test['species'], pred, rownames=['Actual species'], colnames=['Predicted species'])
```

```python
# evaluate model accuracy

from sklearn.metrics import accuracy_score, classification_report

print('Accuracy Score: ', accuracy_score(test['species'], pred))
print('Classification Report: ', classification_report(test['species'], pred))
```

```python
# plot feature importance

importances = clf.feature_importances_
feat_importance_series = pd.Series(importances, index=features)
feat_importance_series.nlargest(4).plot(kind='barh', color='teal')

plt.title("Top Feature Importances in Iris Classification")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.tight_layout()
plt.show()
```
